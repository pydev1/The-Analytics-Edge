```python
# Problem 1.1 - Creating Our First Model
climate_data=pd.read_csv("climate_change.csv")
```

```python
climate_data.info()
```

```python
# split the data into a training set, consisting of all the observations up to and including 2006, and a testing set \
# consisting of the remaining years. 
train_data = climate_data[climate_data["Year"] <= 2006]
```

```python
train_data.describe()
```

```python
test_data = climate_data[climate_data["Year"] > 2006]
```

```python
test_data.describe()
```

```python
# Problem 2.1 - Understanding the Model
X = sm.add_constant(train_data[["MEI","CO2", "CH4", "N2O", "CFC-11", "CFC-12", "TSI", "Aerosols"]])
```

```python
# build a linear regression model to predict the dependent variable Temp
regTemp = sm.OLS(train_data["Temp"], X).fit()
```

```python
# Problem 1.1 - Creating Our First Model
# Question 1: find model R2 (the "Multiple R-squared" value):
# Question 2: Which variables are significant in the model? We will consider a variable signficant only if the p-value \
# is below 0.05.

# R2 can be checked via summary of model
regTemp.summary()
```

```python
# Problem 2.2 - Understanding the Model
# Compute the correlations between all the variables in the training set. 
# Question 1: Which of the following independent variables is N2O highly correlated with(absolute correlation greater than 0.7)?
# Options: MEI, CO2, CH4, CFC-11, CFC-12, Aerosols, TSI
# Question 2: Which of the following independent variables is CFC.11 highly correlated with? Select all that apply.
# Options: MEI, CO2, CH4, N2O, CFC-12, Aerosols, TSI

# correlation can be check by calling method corr on dataframe
train_data.corr()
```

```python
# Problem 3 - Simplifying the Model
# Given that the correlations are so high, let us focus on the N2O variable and build a model with only MEI, TSI, Aerosols \
# and N2O as independent variables. 
X2 = sm.add_constant(train_data[["N2O", "MEI", "TSI","Aerosols"]])
regTemp2 = sm.OLS(train_data["Temp"], X2).fit()
regTemp2.summary()
```
