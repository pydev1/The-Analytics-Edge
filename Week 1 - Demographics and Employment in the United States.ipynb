```python
# Problem 3.4 - Integrating Metropolitan Area Data
# Which metropolitan area has the highest proportion of interviewees of Hispanic ethnicity? 

CPS.groupby("MetroArea")["Hispanic"].mean().sort_values(ascending=False).head(1)
```

```python
# Problem 3.5 - Integrating Metropolitan Area Data
# determine the number of metropolitan areas in the United States from which at least 20% of interviewees are Asian.
# 1st way of solving this problem
(CPS.MetroArea[CPS["Race"] == "Asian"].value_counts() * 100 / CPS.MetroArea.value_counts()).sort_values(ascending=False).head(5)
```

```python
# 2nd way of solving this problem 3.5

temp_df = CPS.groupby("MetroArea")["Race"].value_counts(normalize=True)
temp_df.loc[temp_df.index.get_level_values("Race") == "Asian"].mul(100).astype(int).astype(str) + " %"
```

```python
# 3rd way of solving this problem 3.5
df_asian = CPS.groupby("MetroArea")["Race"].value_counts(normalize=True).astype("float").reset_index(name='count')
df_asian[df_asian["Race"] == "Asian"].sort_values(by="count", ascending=False).head(5)
```

```python
# Problem 3.6 - Integrating Metropolitan Area Data
#determine which metropolitan area has the smallest proportion of interviewees who have received no high school diploma.
df_edu = CPS.groupby("MetroArea")["Education"].value_counts(normalize=True).astype("float").reset_index(name="Count")
df_edu[df_edu["Education"]=="No high school diploma"].sort_values(by="Count").head(1)
```
